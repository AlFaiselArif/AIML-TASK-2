```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#Import and load the dataset
df=pd.read_csv('C:/Users/alfai/Downloads/movie_genre_classification_final.csv')
df.head()
```

```python
#Generate summary statistics (mean, median, std, etc)
print('Summary Statistics')
df.describe()
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
#Create histograms and boxplots for numeric features
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols].hist(figsize=(15, 10), bins=20, color='lightgreen')
plt.suptitle("Histograms for Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# 5. Boxplots for Numeric Columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(3, 3, i+1)
    sns.boxplot(data=df, x=col, color='orange')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
#Use pairplot/correlation matrix for feature relationships
plt.figure(figsize=(10, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
df[numeric_cols].skew()
```

```python
#Pairplot for Selected Features
selected = ['Duration', 'Votes', 'Budget_USD', 'BoxOffice_USD']
sns.pairplot(df[selected], hue='Duration')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
